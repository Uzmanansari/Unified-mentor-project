```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Bird Strikes data.xlsx - Bird Strikes.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Function to save plots
def save_plot(fig, filename):
    fig.savefig(filename)
```

```python
#  Bar Plot of Bird Strikes by Airport
top_airports = data['Airport: Name'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_airports.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Airports by Number of Bird Strikes')
plt.xlabel('Airport')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()

```

```python

#  Bar Plot of Bird Strikes by Wildlife Species
top_species = data['Wildlife: Species'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_species.plot(kind='bar', color='lightblue')
plt.title('Top 10 Wildlife Species Involved in Bird Strikes')
plt.xlabel('Species')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

```python
# Case Study 1: Visuals Depicting the Number of Bird Strikes
fig, ax = plt.subplots(figsize=(10, 6))
data['FlightDate'] = pd.to_datetime(data['FlightDate'])
data['Year'] = data['FlightDate'].dt.year
yearly_strikes = data.groupby('Year').size()
yearly_strikes.plot(kind='bar', ax=ax)
ax.set_title('Number of Bird Strikes Per Year')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Strikes')
plt.xticks(rotation=45)
save_plot(fig, 'Number_of_Bird_Strikes_Per_Year.png')
plt.show()
```

```python
# Case Study 2: Yearly Analysis & Bird Strikes in the US
fig, ax = plt.subplots(figsize=(10, 6))
yearly_strikes.plot(kind='line', ax=ax)
ax.set_title('Yearly Analysis of Bird Strikes in the US')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Strikes')
save_plot(fig, 'Yearly_Analysis_Bird_Strikes_US.png')
plt.grid(axis='y')
plt.show()
```

```python
# Case Study 3: Top 10 US Airlines in terms of having encountered bird strikes
top_airlines = data['Aircraft: Airline/Operator'].value_counts().head(10)
fig, ax = plt.subplots(figsize=(10, 6))
top_airlines.plot(kind='bar', ax=ax)
ax.set_title('Top 10 US Airlines with Most Bird Strikes')
ax.set_xlabel('Airline')
ax.set_ylabel('Number of Strikes')
plt.xticks(rotation=45)
save_plot(fig, 'Top_10_US_Airlines_Bird_Strikes.png')
plt.show()
```

```python
# Case Study 4: Airports with most incidents of bird strikes â€“ Top 50
top_airports = data['Airport: Name'].value_counts().head(50)
fig, ax = plt.subplots(figsize=(10, 6))
top_airports.plot(kind='bar', ax=ax)
ax.set_title('Top 50 Airports with Most Bird Strikes')
ax.set_xlabel('Airport')
ax.set_ylabel('Wildlife: Number Struck Actual')
plt.xticks(rotation=90)
save_plot(fig, 'Top_50_Airports_Bird_Strikes.png')
plt.show()
```

```python
# Case Study 5: Yearly Cost Incurred due to Bird Strikes
data['Year'] = data['FlightDate'].dt.year

data['Cost: Total $'] = pd.to_numeric(data['Cost: Total $'], errors='coerce')
yearly_cost = data.groupby('Year')['Cost: Total $'].sum()
fig, ax = plt.subplots(figsize=(10, 6))
yearly_cost.plot(kind='line', marker='o',color='green')
ax.set_title('Yearly Cost Incurred Due to Bird Strikes')
ax.set_xlabel('Year')
ax.set_ylabel('Total Cost (USD)')
ax.grid(True)
plt.xticks(rotation=45)
save_plot(fig, 'Yearly_Cost_Bird_Strikes.png')
plt.show()
```

```python
# Count the number of bird strikes for each phase of flight
phase_counts = data['When: Phase of flight'].value_counts()

# Plot the distribution of bird strikes across different phases of flight
fig, ax= plt.subplots(figsize=(10, 6))
phase_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Bird Strikes Across Phases of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.grid(axis='y')
save_plot(fig, 'Distribution of Bird Strikes Across Phases of Flight.png')
plt.show()
```

```python
# Case Study 7: Altitude of airplanes at the time of strike
fig, ax = plt.subplots(figsize=(10, 6))
data['Feet above ground'] = pd.to_numeric(data['Feet above ground'], errors='coerce')
altitude_strikes = data['Altitude bin'].dropna()
sns.histplot(altitude_strikes, bins=25, kde=True, color='lightblue',ax=ax)
ax.set_title('Altitude Distribution of Bird Strikes')
ax.set_xlabel('Altitude (feet)')
ax.set_ylabel('Number of Strikes')
save_plot(fig, 'Altitude_Bird_Strikes.png')
plt.grid(axis='y')
plt.show()
```

```python
# Case Study 8: Phase of flight at the time of the strike
fig, ax = plt.subplots(figsize=(10, 6))
phase_strikes = data['When: Phase of flight'].value_counts()
phase_strikes.plot(kind='bar', ax=ax)
ax.set_title('Flight Phase During Bird Strikes')
ax.set_xlabel('When: Phase of flight')
ax.set_ylabel('Wildlife: Number Struck Actual')
plt.xticks(rotation=45)
save_plot(fig, 'Flight_Phase_Bird_Strikes.png')
plt.show()
```

```python
# Case Study 9: Average Altitude of the airplanes in different phases at the time of strike
fig, ax = plt.subplots(figsize=(10, 6))
avg_altitude_phase = data.groupby('When: Phase of flight')['Feet above ground'].mean().dropna()
avg_altitude_phase.plot(kind='bar', ax=ax)
ax.set_title('Average Altitude by Flight Phase During Bird Strikes')
ax.set_xlabel('Flight Phase')
ax.set_ylabel('Average Altitude (feet)')
plt.xticks(rotation=45)
save_plot(fig, 'Average_Altitude_By_Flight_Phase.png')
plt.show()
```

```python
# Case Study 10: Effect of Bird Strikes & Impact on Flight
effect_counts = data['Effect: Impact to flight'].value_counts()

# Plot the distribution of effects of bird strikes
fig, ax = plt.subplots(figsize=(12, 6))
effect_counts.plot(kind='bar', color='skyblue')
plt.title('Effect of Bird Strikes')
plt.xlabel('Effect')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
save_plot(fig, 'Effect of Bird Strikes.png')
plt.grid(axis='y')

fig, ax = plt.subplots(figsize=(10, 6))
impact_strikes = data['Effect: Indicated Damage'].value_counts()
impact_strikes.plot(kind='bar', ax=ax)
ax.set_title('Impact of Bird Strikes on Flight')
ax.set_xlabel('Impact')
ax.set_ylabel('Number of Strikes')
plt.xticks(rotation=45)
save_plot(fig, 'Impact_of_Bird_Strikes.png')
plt.show()
```

```python
# Case Study 11: Effect of Strike at Different Altitude
import pandas as pd
import matplotlib.pyplot as plt
# Convert 'Feet above ground' to numeric, coercing errors to NaN
data['Feet above ground'] = pd.to_numeric(data['Feet above ground'], errors='coerce')

# Filter out rows where altitude information is missing
data_with_altitude = data[data['Feet above ground'].notnull()]

# Define altitude bins and labels
altitude_bins = [0, 1000, 5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, float('inf')]
labels = [
    '0-1000', '1001-5000', '5001-10000', 
    '10001-15000', '15001-20000', '20001-25000', 
    '25001-30000', '30001-35000', '35001-40000', '40000+'
]

# Bin the altitude data
data_with_altitude['Altitude Bin'] = pd.cut(data_with_altitude['Feet above ground'], bins=altitude_bins, labels=labels, right=False)

# Count the number of occurrences for each effect of bird strikes in each altitude bin
effect_counts_by_altitude = data_with_altitude.groupby(['Altitude Bin', 'Effect: Impact to flight']).size().unstack().fillna(0)

# Plot the distribution of effects of bird strikes at different altitudes
fig, ax = plt.subplots(figsize=(10, 6))
effect_counts_by_altitude.plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Effect of Bird Strikes at Different Altitudes')
ax.set_xlabel('Altitude Bin')
ax.set_ylabel('Number of Bird Strikes')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.legend(title='Effect')

# Save the plot as a PNG file
save_plot(fig,'Effect of Strike at Different Altitude.png')

# Show the plot
plt.show()
```

```python
# Case Study 12: Were Pilots Informed? & Prior Warning and Effect of Strike Relation

# rows where information about pilot warning and effect of strike is missing
filtered_data = data[(data['Pilot warned of birds or wildlife?'].notnull()) & (data['Effect: Impact to flight'].notnull())]

# number of occurrences for each combination of pilot warning and effect of strike
warning_effect_counts = filtered_data.groupby(['Pilot warned of birds or wildlife?', 'Effect: Impact to flight']).size().unstack(fill_value=0)

# distribution of pilot warning and effect of strike
fig, ax = plt.subplots(figsize=(10, 6))
warning_effect_counts.plot(kind='bar', stacked=True, ax=ax)
plt.title('Relation between Pilot Warning and Effect of Bird Strikes')
plt.xlabel('Pilot Warned Prior to Strike')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.legend(title='Effect of Strike,Indicated Damage')
plt.grid(axis='y')
save_plot(fig, 'Pilots_Informed_Bird_Strikes.png')
plt.tight_layout()
plt.show()
```
