```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
adventureworks_path = 'AdventureWorks_Database.xlsx'
budget_path = 'Budget.xlsx'
```

```python
# Load necessary sheets
df_calendar = pd.read_excel(adventureworks_path, sheet_name='Calender')
df_customers = pd.read_excel(adventureworks_path, sheet_name='Customers')
df_product = pd.read_excel(adventureworks_path, sheet_name='Product')
df_territory = pd.read_excel(adventureworks_path, sheet_name='Territory')
df_sales = pd.read_excel(adventureworks_path, sheet_name='Sales')
df_budget = pd.read_excel(budget_path, sheet_name='Budget', skiprows=3)
```

```python
# Clean Budget Data
df_budget.columns = ['Category', 'Subcategory', 'ProductName', 'ProductKey',
                     'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Total']
df_budget = df_budget.dropna(subset=['ProductKey'])
df_budget['ProductKey'] = pd.to_numeric(df_budget['ProductKey'], errors='coerce')
```

```python
# Merge Sales and Budget
df_sales_budget = pd.merge(df_sales, df_budget, on='ProductKey', how='left')
```

```python
df_sales_budget
```

```python
# Enrich with Customer and Territory Data
df_sales_budget = pd.merge(df_sales_budget, df_customers, on='CustomerKey', how='left')
df_sales_budget = pd.merge(df_sales_budget, df_territory, on='SalesTerritoryKey', how='left')
```

```python
# Handle missing data
df_sales_budget.fillna(0, inplace=True)
```

```python
df_sales_budget
```

```python
# Calculate Variance and Profit Metrics
df_sales_budget['TotalBudget'] = df_sales_budget[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum(axis=1)
df_sales_budget['Variance'] = df_sales_budget['TotalBudget'] - df_sales_budget['UnitPrice'] * df_sales_budget['OrderQuantity']
df_sales_budget['Profit'] = (df_sales_budget['UnitPrice'] - df_sales_budget['StandardCost']) * df_sales_budget['OrderQuantity']
```

```python
# Key Metrics
sales_summary = df_sales_budget.groupby('Category')[['UnitPrice', 'Profit', 'Variance']].sum()
customer_summary = df_sales_budget.groupby('Gender')[['Profit']].mean()
```

```python
# Visualization
sns.set(style="whitegrid")
```

```python
# Sales by Category
plt.figure(figsize=(10, 6))
sales_summary['UnitPrice'].plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.show()
```

```python
# Profit by Gender
plt.figure(figsize=(8, 5))
customer_summary['Profit'].plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Average Profit by Customer Gender')
plt.xlabel('Gender')
plt.ylabel('Profit ($)')
plt.show()
```

```python
# Variance Analysis
plt.figure(figsize=(10, 6))
sns.histplot(df_sales_budget['Variance'], bins=30, kde=True, color='orange')
plt.title('Variance Distribution (Budget vs Actual)')
plt.xlabel('Variance ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyzing Unit Price Distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='UnitPrice', data=df_sales_budget, palette='coolwarm')
plt.title('Unit Price Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Unit Price ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 5 Selling Products
top_products = df_sales_budget.groupby('ProductName')['OrderQuantity'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Top 5 Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

```python

# Profit by Country
country_profit = df_sales_budget.groupby('SalesTerritoryKey')['Profit'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
country_profit.plot(kind='bar', color='green', edgecolor='black')
plt.title('Profit by Country')
plt.xlabel('Country')
plt.ylabel('Total Profit ($)')
plt.xticks(rotation=45)
plt.show()
```

```python

# Profitability Analysis (Profit/Sales by Country)
df_sales_budget['Profitability'] = df_sales_budget['Profit'] / (df_sales_budget['UnitPrice'] * df_sales_budget['OrderQuantity'] + 1e-5) * 100
profitability_country = df_sales_budget.groupby('SalesTerritoryKey')['Profitability'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
profitability_country.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Profitability by Country')
plt.xlabel('Country')
plt.ylabel('Profitability (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Save Processed Data for Dashboard
processed_data_path = 'Processed_Sales_Budget.csv'
df_sales_budget.to_csv(processed_data_path, index=False)
print(f"Processed data saved to {processed_data_path}")
```
