```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('Heart Disease data.csv')

# Basic info and first look at the data
print(data.info())
print(data.describe())
print(data.head())
```

```python
# Checking for missing values
print(data.isnull().sum())

# Drop or fill missing values (depending on the situation)
data = data.dropna()  # Dropping rows with missing values for now

# Removing duplicates
data = data.drop_duplicates()
```

```python
# Plotting the distribution of heart disease cases (assuming the column 'target' exists)
sns.countplot(x='target', data=data)
plt.title('Distribution of Heart Disease Cases')
plt.show()
```

```python
# Assuming 'gender' column exists (1 = Male, 0 = Female)
sns.countplot(x='target', hue='gender', data=data)
plt.title('Heart Disease by Gender')
plt.show()
```

```python
# Plotting heart disease prevalence across different ages
plt.figure(figsize=(10,6))
sns.histplot(data=data, x='age', hue='target', kde=True, multiple="stack")
plt.title('Heart Disease by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation between features and heart disease diagnosis
corr = data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Features')
plt.show()
```
